```python
recent_period = (
    sts[["Year", "Quarter"]]
    .drop_duplicates()
    .sort_values(["Year", "Quarter"], ascending=[False, False])
    .iloc[0]
)

recent_year = recent_period["Year"]
recent_quarter = recent_period["Quarter"]
```

```python
recent_data = sts[
    (sts["Year"] == recent_year) &
    (sts["Quarter"] == recent_quarter)
]
```

```python
txn_distribution = (
    recent_data.groupby(["State", "Transaction Type"])["Transactions"]
    .sum()
    .reset_index()
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))

for txn_type in txn_distribution["Transaction Type"].unique():
    subset = txn_distribution[txn_distribution["Transaction Type"] == txn_type]
    plt.bar(subset["State"], subset["Transactions"], label=txn_type)

plt.xlabel("State")
plt.ylabel("Transactions")
plt.title(f"Transaction Type Distribution ({recent_year} {recent_quarter})")

plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Aggregate data (already done earlier)
txn_distribution = (
    recent_data.groupby(["State", "Transaction Type"])["Transactions"]
    .sum()
    .reset_index()
)

states = txn_distribution["State"].unique()
txn_types = txn_distribution["Transaction Type"].unique()

x = np.arange(len(states))   # Position of states on X-axis
width = 0.2                  # Width of each bar

plt.figure(figsize=(16,8))

for i, txn_type in enumerate(txn_types):
    subset = txn_distribution[
        txn_distribution["Transaction Type"] == txn_type
    ]
    
    plt.bar(
        x + i * width,                      # Shift bars sideways
        subset["Transactions"],
        width=width,
        label=txn_type
    )

plt.xlabel("State")
plt.ylabel("Transactins")
plt.title("Transaction Type Distribution (Grouped)")

plt.xticks(x + width, states, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
trend_df = (
    STU.groupby(["Year", "Quarter"])
      .agg(
          Total_Transactions=("Transactions","sum"),
          Total_Amount=("Amount (INR)","sum")
      )
      .reset_index()
      .sort_values(["Year", "Quarter"])
)

print(trend_df)
```

```python
trend_df["Time"] = (
    trend_df["Year"].astype(str) + "-Q" +
    trend_df["Quarter"].astype(str)
)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    trend_df["Time"],
    trend_df["Total_Transactions"],
    marker='o',
    label="Transactions"
)

plt.plot(
    trend_df["Time"],
    trend_df["Total_Amount"],
    marker='s',
    label="Amount (INR)"
)

plt.xlabel("Time (Year-Quarter)")
plt.ylabel("Value")
plt.title("Transaction Trends Over Time")

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

ax1 = plt.gca()
ax1.plot(trend_df["Time"], trend_df["Total_Transactions"], marker='o',color='Blue', label="Transactions")
ax1.set_ylabel("Transactions")

ax2 = ax1.twinx()
ax2.plot(trend_df["Time"], trend_df["Total_Amount"], marker='s',color='Orange', label="Amount (INR)")
ax2.set_ylabel("Amount (INR)")

plt.xlabel("Time (Year-Quarter)")
plt.title("Transaction Trends Over Time")

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
